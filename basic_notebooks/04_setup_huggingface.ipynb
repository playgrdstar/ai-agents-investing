## Step 1: Create HuggingFace Account

1. Go to [huggingface.co/join](https://huggingface.co/join)
2. Sign up with email, Google, or GitHub
3. Verify your email address

## Step 2: Get Your Access Token

1. Go to [huggingface.co/settings/tokens](https://huggingface.co/settings/tokens)
2. Click **"Create new token"**
3. Name it (e.g., "Finance Agent Course")
4. Select **"Read"** permission (that's all you need)
5. Click **"Generate"**
6. **Copy the token**

```
Format: hf_xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx
```

## Step 3: Set Your Token

Run this cell and paste your token when prompted:

````python
import os
from getpass import getpass

# Prompt for token (hidden input)
token = getpass("Enter your HuggingFace token: ")
os.environ["HF_TOKEN"] = token

print("âœ… Token set!")
print(f"   Token format: {token[:6]}...{token[-4:]}")
````

## Step 5: Test Your Connection

````python
from huggingface_hub import InferenceClient

client = InferenceClient(
    api_key=os.environ["HF_TOKEN"],
)

# Using the recommended model for the course
completion = client.chat.completions.create(
    model="Qwen/Qwen2.5-Coder-32B-Instruct",
    messages=[
        {
            "role": "user",
            "content": "Say 'Hello, Finance Agent!' and nothing else."
        }
    ],
    max_tokens=100
)

print(completion.choices[0].message.content)
````

## Step 6: Test with Chat Completion API

````python
# Chat completion with system message (OpenAI-compatible API)
response = client.chat.completions.create(
    model="Qwen/Qwen2.5-Coder-32B-Instruct",
    messages=[
        {"role": "system", "content": "You are a helpful finance assistant. Be concise."},
        {"role": "user", "content": "What is the P/E ratio and what does a high value mean?"}
    ],
    max_tokens=200
)

print(response.choices[0].message.content)
````

````python
from smolagents import CodeAgent, InferenceClientModel

# This is the recommended model for the FREE tier
model = InferenceClientModel(model_id="Qwen/Qwen2.5-Coder-32B-Instruct")
agent = CodeAgent(tools=[], model=model)

result = agent.run("Calculate the monthly payment on a $200,000 mortgage at 6.5% for 30 years.")
print(result)
````
