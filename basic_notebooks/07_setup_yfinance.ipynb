## Step 2: Test Your Connection - Current Price

```python
import yfinance as yf

# Get Apple stock
ticker = yf.Ticker("AAPL")

# Get current price info
info = ticker.info

print("âœ… Connection successful!")
print(f"   Company: {info.get('longName', 'N/A')}")
print(f"   Current Price: ${info.get('currentPrice', info.get('regularMarketPrice', 'N/A'))}")
print(f"   Market Cap: ${info.get('marketCap', 0)/1e9:.1f}B")
print(f"   52-Week High: ${info.get('fiftyTwoWeekHigh', 'N/A')}")
print(f"   52-Week Low: ${info.get('fiftyTwoWeekLow', 'N/A')}")
```

## Step 3: Get Historical Data

```python
# Get historical prices
df = ticker.history(period="1mo")  # Last month

print("âœ… Historical data retrieved!")
print(f"   Period: {df.index[0].date()} to {df.index[-1].date()}")
print(f"   Records: {len(df)}")
print(f"\nLast 5 days:")
print(df[["Open", "High", "Low", "Close", "Volume"]].tail())
```

## Step 4: Different Time Periods

```python
# Various period options
periods = {
    "1d": "1 day (intraday)",
    "5d": "5 days",
    "1mo": "1 month",
    "3mo": "3 months",
    "6mo": "6 months",
    "1y": "1 year",
    "2y": "2 years",
    "5y": "5 years",
    "10y": "10 years",
    "ytd": "Year to date",
    "max": "Maximum available"
}

print("Available periods:")
for period, description in periods.items():
    print(f"  â€¢ {period:5} â†’ {description}")

# Example: Get 5 years of data
df_5y = ticker.history(period="5y")
print(f"\n5-year data: {len(df_5y)} trading days")
```

## Step 6: Multiple Tickers at Once

```python
# Download multiple tickers
tickers = ["AAPL", "MSFT", "GOOGL", "NVDA"]
data = yf.download(tickers, period="1mo", progress=False)

print("âœ… Multiple tickers downloaded!")
print(f"   Tickers: {tickers}")
print(f"   Records: {len(data)}")
print(f"\nClosing prices (last 5 days):")
print(data["Close"].tail())
```

## Step 7: Financial Statements

```python
# Get financial statements
aapl = yf.Ticker("AAPL")

print("=" * 50)
print("INCOME STATEMENT (Annual)")
print("=" * 50)
income = aapl.income_stmt
if not income.empty:
    # Show key metrics
    key_rows = ["Total Revenue", "Gross Profit", "Operating Income", "Net Income"]
    for row in key_rows:
        if row in income.index:
            latest = income.loc[row].iloc[0]
            print(f"  {row}: ${latest/1e9:.1f}B")
else:
    print("  Not available")

print("\n" + "=" * 50)
print("BALANCE SHEET (Latest)")
print("=" * 50)
balance = aapl.balance_sheet
if not balance.empty:
    key_rows = ["Total Assets", "Total Liabilities Net Minority Interest", "Total Equity Gross Minority Interest"]
    for row in key_rows:
        if row in balance.index:
            latest = balance.loc[row].iloc[0]
            print(f"  {row}: ${latest/1e9:.1f}B")
```

## Step 8: Create Tools for smolagents

```python
!pip install -q smolagents
```

```python
from smolagents import tool

@tool
def get_stock_price(ticker: str) -> dict:
    """Get current stock price and key metrics using yfinance.

    Args:
        ticker: Stock symbol (e.g., 'AAPL', 'MSFT', 'GOOGL')

    Returns:
        Dictionary with price, market cap, and 52-week range
    """
    import yfinance as yf

    stock = yf.Ticker(ticker.upper())
    info = stock.info

    return {
        "ticker": ticker.upper(),
        "name": info.get("longName", "N/A"),
        "price": info.get("currentPrice", info.get("regularMarketPrice")),
        "market_cap": info.get("marketCap"),
        "52_week_high": info.get("fiftyTwoWeekHigh"),
        "52_week_low": info.get("fiftyTwoWeekLow"),
        "pe_ratio": info.get("trailingPE"),
        "source": "Yahoo Finance"
    }

@tool
def get_historical_prices(ticker: str, period: str = "1mo") -> str:
    """Get historical stock prices.

    Args:
        ticker: Stock symbol (e.g., 'AAPL', 'MSFT')
        period: Time period - '1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', 'max'

    Returns:
        String summary of historical prices
    """
    import yfinance as yf

    stock = yf.Ticker(ticker.upper())
    df = stock.history(period=period)

    if df.empty:
        return f"No data found for {ticker}"

    return {
        "ticker": ticker.upper(),
        "period": period,
        "start_date": str(df.index[0].date()),
        "end_date": str(df.index[-1].date()),
        "start_price": round(df["Close"].iloc[0], 2),
        "end_price": round(df["Close"].iloc[-1], 2),
        "high": round(df["High"].max(), 2),
        "low": round(df["Low"].min(), 2),
        "avg_volume": int(df["Volume"].mean()),
        "trading_days": len(df)
    }

# Test the tools
print("Tool 1 - Current Price:")
print(get_stock_price("NVDA"))

print("\nTool 2 - Historical:")
print(get_historical_prices("NVDA", "3mo"))
```

## Available Data Reference

| Data Type | Method | Notes |
|-----------|--------|-------|
| Current price | `ticker.info` | Real-time (slight delay) |
| Historical prices | `ticker.history()` | Daily, weekly, monthly |
| Intraday | `ticker.history(period='1d', interval='1m')` | 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h |
| Dividends | `ticker.dividends` | Historical dividends |
| Splits | `ticker.splits` | Stock splits |
| Income Statement | `ticker.income_stmt` | Annual/quarterly |
| Balance Sheet | `ticker.balance_sheet` | Annual/quarterly |
| Cash Flow | `ticker.cashflow` | Annual/quarterly |
| Options | `ticker.options` | Available expiration dates |
| Recommendations | `ticker.recommendations` | Analyst recommendations |
